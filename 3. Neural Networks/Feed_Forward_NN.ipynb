```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('ex3data1.mat')

X = data['X']
y = data['y']
```

```python
fig, axes = plt.subplots(10, 10, figsize = (10, 10))
for i in range(10):
    for j in range(10):
        axes[i, j].imshow(X[np.random.randint(X.shape[0])].reshape((20, 20), order = 'F'), cmap = 'hot')
        axes[i, j].axis('off')
```

```python
weights = loadmat('ex3weights.mat')

theta1 = weights['Theta1']
theta2 = weights['Theta2']
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def feedForwardPropagate(X, theta1, theta2):
        m = len(X)
        ones = np.ones((m, 1))
        X = np.hstack((ones, X))
        a1 = sigmoid(np.dot(X, theta1.T))
        a1 = np.hstack((ones, a1))
        h = np.dot(a1, theta2.T)
        return h
```

Input layer size = 400
Hidden layer size = 25
Number of Labels = 10

```python
h = feedForwardPropagate(X, theta1, theta2)
```

```python
def predict(h):
    return np.argmax(h, axis = 1) + 1

pred = predict(h)
acc = np.mean(pred == y.flatten()) * 100

print('Accuracy = '+ str(acc))
```
